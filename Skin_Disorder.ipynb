## DOMAIN EXPERTISE

### Clinical Attributes

##### 1. Erythema
- Erythema or redness of the skin caused by injury or another inflammation-causing condition. Often presenting as a rash, erythema can be caused by environmental factors, infection, or overexposure to the sun (ie, sunburn).
- Values range from 0 to 3.
##### 2. Scaling
- Scales are a visible peeling or flaking of outer skin layers.
- Values range from 0 to 3.
##### 3. Definite Borders
- Uneven (irregular) borders, values range from 0 to 3.
##### 4. Itching
- Itching is an irritating sensation that makes you want to scratch your skin.
- Values range from 0 to 3.
##### 5. Koebner Phenomenon
- Koebner phenomenon describes the appearance of new skin lesions of a pre-existing dermatosis on areas of cutaneous injury in otherwise healthy skin.
- Values range from 0 to 3.
##### 6. Polygonal Papules
- Papules and polygonal plaques are shiny, flat-topped and firm on palpation.
- Values range from 0 to 3.
##### 7. Follicular Papules
- A follicular papule occurs when the inflammatory infiltrate accumulates within the epidermis of the follicle or in the perifollicular dermis.
- Values range from 0 to 3.
##### 8. Oral Mucosal Involvement
- The oral mucosa is the mucous membrane lining or “skin” inside of the mouth, including cheeks and lips.
- Values range from 0 to 3.
##### 9. Knee and Elbow Involvement
- Causes a rash with itchy, scaly patches, most commonly on the knees, elbows, trunk and scalp.
- Values range from 0 to 3.
##### 10. Scalp Involvement
- Scales patches, itchy on scalps; Value ranges from 0 to 3.
##### 11. Family History
- Whether diseases are observed in the family.
- Values 1 and 0, 1 if any of these diseases has been observed in the family and 0 if not.

### Histopathological Attributes

##### 12. Melanin Incontinence
- Melanin incontinence (pigmentary incontinence) is the presence of melanin in the superficial dermis, due to the loss of melanin from damaged cells of the basal layer.
- Values ranges from 0 to 3.
##### 13. Eosinophils in the Infiltrate
- Affects specific tissue layers or adnexal structures of the skin.
- Values ranges from 0 to 3.
##### 14. PNL Infiltrate
- Pure Neural Leprosy (PNL) is a rare clinical form of leprosy in which patients do not present with the classical skin lesions but have a high burden of the disability associated with the disease.
- Values ranges from 0 to 3.
##### 15. Fibrosis of the Papillary Dermis
- Fibroplasia of papillary dermis is a frequent histological finding present in dysplastic melanocytic nevi and melanomas.
- Values ranges from 0 to 3.
##### 16. Exocytosis
- Exocytosis is the process by which cells move materials from within the cell into the extracellular fluid.
- Values ranges from 0 to 3.
##### 17. Acanthosis
- Acanthosis nigricans is a velvety, darkening of the skin that usually occurs in intertriginous areas.
- Values ranges from 0 to 3.
##### 18. Hyperkeratosis
- Hyperkeratosis is a condition that causes your skin to thicken in certain places.
- Values ranges from 0 to 3.
##### 19. Parakeratosis
- Parakeratosis refers to incomplete maturation of epidermal keratinocytes, resulting in abnormal retention of nuclei in the stratum corneum.
- Values ranges from 0 to 3.
##### 20. Clubbing of the Rete Ridges
- Clubbed and evenly elongated rete ridges were more common in the Psoriasis group than in the Seboreic dermatitis groups.
- Values ranges from 0 to 3.
##### 21. Elongation of the Rete Ridges
- Psoriasis is the prototype of a group of cutaneous disorders (psoriasiform dermatitides) that show psoriasiform epidermal hyperplasia, defined as regular elongation of the rete ridges with preservation of the rete ridge-dermal papillae pattern.
- Values ranges from 0 to 3.
##### 22. Thinning of the Suprapapillary Epidermis
- This was defined as a thinning of the granular layer at the tips of the papillae.
- Values ranges from 0 to 3.
##### 23. Spongiform Pustule
- Spongiform pustule of Kogoj, due to neutrophils migrations from superficial dermal capillaries to intercellular spaces of the edematous Malpighian layer, leading to keratinocytes degeneration and formation of unilocular pustules.
- Values ranges from 0 to 3.
##### 24. Munro Microabcess
- Munro's microabscess is an abscess (collection of neutrophils) in the stratum corneum of the epidermis due to the infiltration of neutrophils from papillary dermis into the epidermal stratum corneum.
- Values ranges from 0 to 3.
##### 25. Focal Hypergranulosis
- Hypergranulosis is an increased thickness of the stratum granulosum.
- Values ranges from 0 to 3.
##### 26. Disappearance of the Granular Layer
- Parakeratosis is associated with the thinning or loss of the granular layer and is usually seen in diseases of increased cell turnover, whether inflammatory or neoplastic.
- Values ranges from 0 to 3.
##### 27.Vacuolisation and Damage of Basal Layer
- In dermatopathology "vacuolization" often refers specifically to vacuoles in the basal cell-basement membrane zone area, where it is an unspecific sign of disease.
- Values ranges from 0 to 3.
##### 28. Spongiosis
- Spongiosis is mainly intercellular edema (abnormal accumulation of fluid) in the epidermis, and is characteristic of eczematous dermatitis, manifested clinically by intraepidermal vesicles (fluid-containing spaces), "juicy" papules and/or lichenification.
- Values ranges from 0 to 3.
##### 29. Saw-tooth Appearance of Retes
- A 'sawtooth' pattern of the rete can be observed in oral lichen planus but this histologic finding is more common in cutaneous LP.
- Values ranges from 0 to 3.
##### 30. Follicular Horn Plug
- The precise cause or causes of follicular occlusion are unknown. There are genetic, hormonal and environmental influences.
- Values ranges from 0 to 3.
##### 31. Perifollicular Parakeratosis
- Unique epidermal reaction representing an acquired disorder of keratinization.
- Values ranges from 0 to 3.
##### 32. Inflammatory Monoluclear Inflitrate
- An inflammatory component is supported by the mononuclear cell infiltration in the myocardium as well as by the deposition of IgG, complement, and fibrin.
- Values ranges from 0 to 3.
##### 33. Band-like Infiltrate
- A bandlike infiltrate of inflammatory cells in the superficial dermis, parallel to the epidermis.
- Values ranges from 0 to 3.
##### 34. Age
- Represents the age of the patient
##### 35. Class (Target Variable)
- 1: Psoriasis - Psoriasis is a chronic disease characterized by inflammation with periods of exacerbation and remission, with the formation of erythematous plaques or papules with pearlescent-white scales on the skin. 
- 2: Seboreic Dermatitis - Seborrheic Dermatitis is a chronic inflammatory disease characterized by oily yellowish scaly and erythematous scaling patches, which can occur on the face, scalp and body folds, periodically recurs or heals. 
- 3: Lichen Planus - Lichen Planus is a papulosquamous, non-infectious inflammatory disease involving the skin and mucous membranes. It does not show organ involvement and involves skin, nails and mucosal surfaces. It is more common in women.
- 4: Pityriasis Rosea - Pityriasis rosea is a self-limiting, common acute papulosquamous inflammatory disease affecting the trunk and extremities
- 5: Chronic Dermatitis - It is a disease known as eczema in Turkey. Dermatitis types include atopic dermatitis, seborrheic dermatitis, contact dermatitis, allergic contact dermatitis, irritant contact dermatitis, hypereosinophilic dermatitis and other types. 
- 6: Pityriasis Rubra Pilaris - Pityriasis Rubra Pilaris (PRP) is a rare inflammatory disease, the etiology of which is not fully known, and the definitive treatment is still being investigated.

## MODEL BUILDING


### Logistic Regression Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score,roc_auc_score

# Define the features and the target variable
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_new=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4)
```

```python
# Create a logistic regression classifier
lr_clf=LogisticRegression()
lr_clf.fit(X_train,y_train)
y_pred_LR=lr_clf.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred_LR)
print(cm)
```

```python
pd.crosstab(y_test,y_pred_LR)
```

```python
print(classification_report(y_test,y_pred_LR))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier#importing randomforest

rf_clf = RandomForestClassifier(n_estimators=100)#object creation ,taking 100 decision tree in random forest 
rf_clf.fit(X_train,y_train)#training the data
```

```python
y_pred_RF=rf_clf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_RF))
```

### XGBoost

```python
#!pip install xgboost
```

```python
import xgboost as xgb
# Create an XGBoost classifier
xgb_classifier = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# Train the XGBoost classifier
xgb_classifier.fit(X_train, y_train)

# Make predictions on the test set
Y_pred_xgb = xgb_classifier.predict(X_test)
```

```python
print(classification_report(y_test,Y_pred_xgb ))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred_dt = dt.predict(X_test)
print(classification_report(y_test,y_pred_dt))
```

### Support Vector Machine

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
y_pred_svc = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_svc))
```

## PREDICTING THE TYPE OF DISEASE

```python
print("Accuracy of Logistic Regression: ", accuracy_score(y_test,y_pred_LR))
print("Accuracy of Random Forest: ", accuracy_score(y_test,y_pred_RF))
print("Accuracy of XGBoost: ", accuracy_score(y_test,Y_pred_xgb))
print("Accuracy of Decision Tree: ", accuracy_score(y_test,y_pred_dt))
print("Accuracy of Support Vector Machine: ", accuracy_score(y_test,y_pred_svc))
```
